```python

# Hàm xử lý địa chỉ
def split_address_components(df):
    df['Tỉnh/ Thành phố'] = df['Địa chỉ'].str.split(',').str[-1].str.strip()
    df['Quận/ Huyện'] = df['Địa chỉ'].str.split(',').str[-2].str.strip()
    df['Xã/ Phường'] = df['Địa chỉ'].str.split(',').str[-3].str.strip()
    df['Đường'] = df['Địa chỉ'].str.split(',').str[0].str.strip()
    #df.drop(columns=['Địa chỉ'], inplace=True)
    return df
#Hàm loại bỏ từ
def remove_words(df):
    for i in range(df.shape[0]):
      if df.at[i, 'Tỉnh/ Thành phố'] != 'Hồ Chí Minh':
        df['Quận/ Huyện'][i]=df['Quận/ Huyện'][i].replace('Quận ',"").replace("Huyện ","")
        df['Tỉnh/ Thành phố'][i]=df['Tỉnh/ Thành phố'][i].replace("Tỉnh ",'').replace('Thành phố ',"")
      else:
        if any(char.isdigit() for char in df.at[i,'Quận/ Huyện']):
          if df['Quận/ Huyện'][i]== 'Quận 2' or df['Quận/ Huyện'][i]=='Quận 9':  df['Quận/ Huyện'][i]='Thủ Đức'
        else:
          df['Quận/ Huyện'][i]=df['Quận/ Huyện'][i].replace('Quận ',"").replace("Huyện ","")
        df.at[i,'Tỉnh/ Thành phố']='Thành phố Hồ Chí Minh'
      if 'Huế' in df.at[i,'Tỉnh/ Thành phố']: df['Tỉnh/ Thành phố'][i]='Thừa Thiên Huế'
      if 'Bà Rịa Vũng Tàu' in df.at[i,'Tỉnh/ Thành phố']: df['Tỉnh/ Thành phố'][i]= 'Bà Rịa – Vũng Tàu'
      if 'Thành phố' in df.at[i,'Quận/ Huyện']:
        df['Quận/ Huyện'][i]=df['Quận/ Huyện'][i].replace('Thành phố ',"")
      if 'Thị xã' or 'Thị Xã' in df.at[i,'Quận/ Huyện']:
        df['Quận/ Huyện'][i]=df['Quận/ Huyện'][i].replace('Thị xã ',"")
      if 'Phan Rang' in df.at[i,'Quận/ Huyện']: df['Quận/ Huyện'][i]='Phan Rang – Tháp Chàm'
      if 'Biên Hoà' in df.at[i,'Quận/ Huyện']: df['Quận/ Huyện'][i]='Biên Hòa'
    return df
```

```python
#Hàm đối chiếu để lấy về mật độ dân số theo quận/huyện
def add_density_column(df1, den):
    df1["Mật độ dân số (người/km²)"] = 0  # Tạo cột mới và điền giá trị mặc định là 0
    for index, row in df1.iterrows():
        huyen = row["Quận/ Huyện"]
        tinh = row['Tỉnh/ Thành phố']
        matched_row = den[(den["Tên đơn vị hành chính"] == huyen) & (den['Tỉnh/Thành phố'] == tinh)]
        if not matched_row.empty:
            df1.at[index, "Mật độ dân số (người/km²)"] = matched_row.iloc[0]["Mật độ dân số (người/km²)"]
    return df1
def add_GDPD(df1,grdp):
    df1["GRDP(%)"]=0
    for index, row in df1.iterrows():
      tinh = row['Tỉnh/ Thành phố']
      matched_row= grdp[grdp["Tên tỉnh / thành phố"]== tinh]
      if not matched_row.empty:
        df1.at[index, "GRDP(%)"]=matched_row.iloc[0]["Tăng trưởng 2023"]
    return df1
```

```python
#Chuyển đơn vị độ dài
def convert_do_dai(dtf, oldname, newname):
    for i in range(dtf.shape[0]):
      string=dtf.at[i, oldname]
      if string != '---':
        if 'm' in string:
          dtf.at[i, oldname] = pd.to_numeric(string.replace('m', ''), errors='coerce')
        elif 'km' in string:
          pass

    dtf.rename(columns={oldname: newname}, inplace=True)
    return dtf
#Chuyển đơn vị điện tích
def convert_dien_tich(dtf, oldname, newname):
    for i in range(dtf.shape[0]):
        string=dtf.at[i, oldname]
        if string != '---':
            dtf.at[i, oldname] = pd.to_numeric(string.replace(' m2', ''), errors='coerce')
    dtf.rename(columns={oldname: newname}, inplace=True)
    return dtf
#Chuyển giá tiền
def convert_giaca(dtf,oldname,newname,area):
  for i in range(dtf.shape[0]):
    string=dtf.at[i,oldname]
    string=string.replace(',','.')
    if ' ngàn' in string:
      dtf = dtf.drop(i)
    if 'Thỏa thuận' in string:
      dtf = dtf.drop(i)
    if ' tỷ / m2' in string:
      dt=dtf.at[i,area]
      price = pd.to_numeric(string.replace(' tỷ / m2',''),errors='coerce')
      dtf.at[i,oldname]= dt* price
    if ' tỷ' in string:
      dtf.at[i,oldname]=pd.to_numeric(string.replace(' tỷ', ''), errors='coerce')
    if ' triệu / m2' in string:
      dt=dtf.at[i,area]
      price = pd.to_numeric(string.replace(' triệu / m2',''),errors='coerce')
      dtf.at[i,oldname]= dt* price/1000
    if ' triệu' in string:
      price = pd.to_numeric(string.replace(' triệu',''),errors='coerce')
      dtf.at[i,oldname]= price/1000
  dtf=dtf.reset_index(drop=True)
  dtf.rename(columns={oldname: newname}, inplace=True)
  return dtf

def convert_percent(dtf, clum):
    for i in range(dtf.shape[0]):
        value = dtf.at[i, clum]
        if isinstance(value, (int, float)):
            continue
        dtf.at[i, clum] = pd.to_numeric(str(value).replace(',', '.').replace('%', ''))
    return dtf
```
